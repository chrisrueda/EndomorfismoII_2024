## Modelo lineal
Este es un ejemplo de un modelo lineal de aprendizaje.
Este modelo busca relacionar el tamaño de la vivienda con el precio de la misma.

```python
#Añadir librerías usadas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Ahora vamos a añadir los datos:

```python
#conjunto de datos
x = np.array([650, 700, 750, 800, 850]).reshape(-1,1) #metros cuadrados para el tamaño de la vivienda (Variable independiente)
y = np.array([300, 310, 320, 330, 340]) #Precio de la vivienda

# si tengo 100 muestras - cuántas son para entrenamiento? cuántas para validación? 70% 30%    80% -20%
#Dividir conjunto de entrenamiento y validación
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

#Crear el modelo

```python
modelo = LinearRegression()
#Entrenar el modelo
modelo.fit(X_train, y_train) 
```

```python
#Realizar la predicción con los valores de test
prediction = modelo.predict(X_test)
print("Valores estimados: ")
print(prediction)
print("Valores reales: ")
print(y_test)
```

```python
#Visualizar los datos
plt.scatter(X_train, y_train, color='b', label='Datos de entrenamiento')
plt.scatter(X_test, y_test, c='g', label='Datos de validación')
plt.plot(X_test, prediction, c='r', label='Valores predichos')
plt.legend()
plt.show()
```
