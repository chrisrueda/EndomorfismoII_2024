## 1. Importar y visualizar datos

```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt

# Generar nuestros puntos 
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

# Graficar la compuerta xor
plt.scatter(X[:, 0], X[:,1], c=y, cmap='winter')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Problema XOR')
plt.legend()
plt.grid()
plt.show()
```

## Implementación del modelo

```python
# En IA, 2 problemas: 
# Clasificación: Clases o grupos.
# Regresión: Predecir 
from sklearn.svm import SVC

# Generar el modelo
modelo = SVC(kernel='rbf') #modelo parametrizado
# Entrenar el modelo
modelo.fit(X, y) #modelo entrenado

# Clasificar con el modelo entrenado
y_estimada = modelo.predict(X) #modelo entrenado ha determinado
print(f"y estimada: {y_estimada}")
print(f"y real: {y} ")
```

## Visualización de la frontera de decisión del SVM

```python
# Crear malla de puntos
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1

xx, yy, = np.meshgrid(np.arange(x_min, x_max, 0.01),
                      np.arange(y_min, y_max, 0.01))

# Predecir para el gráfico
y_graf = modelo.predict(np.c_[xx.ravel(), yy.ravel()])
y_graf = y_graf.reshape(xx.shape)
# Frontera de decisión
plt.contourf(xx, yy, y_graf, alpha=0.5, cmap = 'winter')
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter')
```

## Prueba con un nuevo valor


```python
X_test = np.array([[1.0, -0.5]])

# Clasificar con el nuevo dato
y_test_pred = modelo.predict(X_test)
print(f"El punto de prueba se ha clasificado como: {y_test_pred}")
```
