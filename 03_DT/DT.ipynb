# Árboles de decisión

## Árbol de decisión para datos aleatorios

```python
# Importar las librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

## Generar los datos aleatorios

```python
np.random.seed(42) #Agregar una semilla para reprod

X = np.random.randn(200,2) # Vector de entrada
y = (X[:,0]**2+X[:,1]**2) <0.5 # Target - Vector de etiquetas
```

## Graficar datos

```python
plt.scatter(X[y==0][:,0], X[y==0][:,1], c='red', label= 'clase 0')
plt.scatter(X[y==1][:,0], X[y==1][:,1], c='blue', label= 'clase 1')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.legend()
plt.show()
```

## Crear el modelo

```python
modelo = DecisionTreeClassifier(max_depth=2, criterion='gini')
modelo.fit(X,y)

#Predecir con el modelo
y_pred = modelo.predict(X)

#Comparar las respuestas
#print(f"y real: {y}")
#print(f"y predicha: {y_pred}")
```

```python
#Graficar el árbol de decisión
from sklearn import tree

plt.figure(figsize=(20,10))

tree.plot_tree(modelo, 
               filled=True, 
               rounded=True, 
               class_names=["clase 0", "clase 1"],
               feature_names=["f1", "f2"]
            )
plt.show()
```
