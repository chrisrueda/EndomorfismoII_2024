# Entrenamiento de los datos

```python
# Matriz X de entrada
X = df.drop(["SP",'ID'], axis=1)
#X=df.drop('ID',axis=1)
# vector y de salida
Y = df["SP"]

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=20) 
```

#### Comprobacion de las matrices

```python
X.head(10)
```

```python
Y.head(10)
```

### Entrenamiento

```python
# Importar modelo random forest. 
rf = RandomForestClassifier(n_estimators=50, random_state=42)
#entrenar modelo 
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test) # variable predichas por el modleo

exactitud = accuracy_score(y_test, y_pred)

print(f"El modelo de Random Forest obtuvo una exactitud de: {exactitud}")
```

### Matriz de confucion

```python
# Matriz de confusion
con_mat=confusion_matrix(y_test,y_pred)
print(f'Matriz de confusion:\n{con_mat}')
#exactitud
exactitud=accuracy_score(y_test,y_pred)
#precision
precision=precision_score(y_test,y_pred)
#sensibilidad
recall=recall_score(y_test,y_pred)

#Especificidad

vn,fp,fn,vp=con_mat.ravel()
especificidad=vn/(vn+fp)
#f1-score
f1=f1_score(y_test,y_pred)

print(f'Exactitud: {exactitud: .2f}\nPresicion:{precision: .2f}\nSensibilidad: {recall: .2f}\nEspecificidad{especificidad: .2f}\nf1-score: {f1: .2f}')
```

###### El proceso es capas de identificar de mejor manera los casos positivos pero no logra predecir bien los casos negativos

```python
# EVALUACION DEL MODELO
print(classification_report(y_test,y_pred))
```

#### Boosting

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(use_label_encoder=False)

#entrenamiento 
xgb.fit(X_train, y_train)
#Prediccion 
y_pred = xgb.predict(X_test)
#Validacion
print(classification_report(y_test, y_pred))
print("Matriz de confusion: ")
print(confusion_matrix(y_test, y_pred))
```
