#### Datos a corregir:
M1-M6, se detalla q -1 es a tiempo y 1 es retraso de un mes, pero no hay descripción para 0 ni -2, 0 tanto 0 como -2 sera imputado.

D1-D6  Al ser una cuenta de credito si es posible que se encuentre en números negativos.

P1-P6 No se presentan casos a modoficar como pagos negativos.

ID y G serán eliminado

EC se eliminaran los nan y 0


```python
# verificar los valores -2
col_dos = ['M1', 'M2', 'M3', 'M4', 'M5', 'M6']
data_sec = data[col_dos]
data_dos = pd.DataFrame(data_sec == -2)
data_zero = pd.DataFrame(data_sec == 0)

for i in data_dos.columns.values.tolist():
    print(data_dos[i].value_counts())
    print(data_zero[i].value_counts())
```

```python
# Imputar los datos
#M1
m1 = data["M1"].value_counts().idxmax()
print(f"La variable más frecuente para M1 es: {m1}")
#Imputar variable
data["M1"].replace(0, -1, inplace=True)
data["M1"].replace(-2, m1, inplace=True)
#M2
m2 = data["M2"].value_counts().idxmax()
print(f"La variable más frecuente para M2 es: {m2}")
#Imputar variable
data["M2"].replace(0, -1, inplace=True)
data["M2"].replace(-2, m2, inplace=True)
#M3
m3 = data["M3"].value_counts().idxmax()
print(f"La variable más frecuente para M3 es: {m3}")
#Imputar variable
data["M3"].replace(0, -1, inplace=True)
data["M3"].replace(-2, m3, inplace=True)
#M4
m4 = data["M4"].value_counts().idxmax()
print(f"La variable más frecuente para M4 es: {m4}")
#Imputar variable
data["M4"].replace(0, -1, inplace=True)
data["M4"].replace(-2, m4, inplace=True)
#M5
m5 = data["M5"].value_counts().idxmax()
print(f"La variable más frecuente para M5 es: {m5}")
#Imputar variable
data["M5"].replace(0, -1, inplace=True)
data["M5"].replace(-2, m5, inplace=True)
#M6
m6 = data["M6"].value_counts().idxmax()
print(f"La variable más frecuente para M6 es: {m6}")
#Imputar variable
data["M6"].replace(0, -1, inplace=True)
data["M6"].replace(-2, m6, inplace=True)
#EC
ec = data["EC"].value_counts().idxmax()
print(f"La variable más frecuente para EC es: {ec}")
#Imputar variable
data["EC"].replace('0', ec, inplace=True)
data["EC"].replace(np.nan, ec, inplace=True)
```

```python
#Se reemplazaran los datos categoricos por númericos
#EC
data["EC"].replace('Casado', 1, inplace=True)
data["EC"].replace('Soltero', 2, inplace=True)
data["EC"].replace('Otros', 3, inplace=True)
#SP
data["SP"].replace('Si', 1, inplace=True)
data["SP"].replace('No', 0, inplace=True)
```

```python
data.head(15)
```

### Generación de modelo

```python
#Matriz de entrada
x = data.drop(['G','SP'], axis=1)
#Target - Vector de salida
y = data['SP']

#Generar el conjunto de entremamiento y validación
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.3, random_state=20)

#Creal modelo
modelo_rf = RandomForestClassifier()
#Entrenar el modelo
modelo_rf.fit(x_train, y_train)
#Validar modelo
y_pred = modelo_rf.predict(x_test)

#Metricas

# Matriz de confusión
con_mat = confusion_matrix(y_test, y_pred)
print(f"Matriz de confusión:\n {con_mat}")

#Exactitud
exactitud=accuracy_score(y_test, y_pred)
print(f"Exactitud: {exactitud:.2f}")

#Precisión
precision = precision_score(y_test, y_pred, average="binary", pos_label=0)
print(f"precision: {precision:.2f}")

#Sensibilidad
recall = recall_score(y_test, y_pred, average="binary", pos_label=0)
print(f"recall: {recall:.2f}")

#Especificidad
vn, fp, fn, vp = con_mat.ravel()
especificidad = vn / (vn + fp)
print(f"especificidad: {especificidad:.2f}")

#f1-score
f1 = f1_score(y_test, y_pred, average="binary", pos_label=0)
print(f"f1: {f1:.2f}")
```

### Modelo XGBoost

```python
xgb = XGBClassifier(use_label_encoder=False)
#Entrenamiento
xgb.fit(x_train, y_train)
#predicción
y_predboost = xgb.predict(x_test)
#Validación
print(classification_report(y_test, y_predboost))
print("Matriz de confusión")
print(confusion_matrix(y_test, y_predboost))
```

```python
#Exactitud
exactitudXG=accuracy_score(y_test, y_predboost)
#Precisión
precisionXG = precision_score(y_test, y_predboost)
#Sensibilidad
recallXG = recall_score(y_test, y_predboost)

#Especificidad
vn, fp, fn, vp = con_mat.ravel()
especificidadXG = vn / (vn + fp)
#f1-score
f1XG = f1_score(y_test, y_predboost)

print(f"Exactitud: {exactitudXG:.2f}\nPrecisión: {precisionXG:.2f}\nSensibilidad: {recallXG:.2f}\nEspecificidad: {especificidadXG:.2f}\nf1-Score: {f1XG:.2f}")
```
