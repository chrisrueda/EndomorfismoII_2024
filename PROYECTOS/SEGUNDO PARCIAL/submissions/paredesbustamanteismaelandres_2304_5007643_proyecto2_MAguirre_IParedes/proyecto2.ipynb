### 1. Analizar datos

```python
import pandas as pd
import numpy as np

url="https://raw.githubusercontent.com/chrisrueda/EndomorfismoII_2024/main/PROYECTOS/SEGUNDO%20PARCIAL/mora_cdto.csv"

df=pd.read_csv(url)

df.head(10)
```

### Análisis estadístico simple

```python
df.info()
```

Se observa que el dataframe tiene diferentes tipos de datos y datos faltantes, por lo que se tendrá que tratarlos

```python
df.describe()
```

Hay que modificar G, EC por datos faltantes (hay que imputarlos) y al no ser datos numéricos pero también hay que modificar SP al no ser datos numéricos

### Analizar el balance de los datos

```python
df['SP'].unique()

print("Próximo pago en mora:")
df['SP'].value_counts()
```

### Datos Faltantes

```python
datos_falt=df.isnull()

for carac in datos_falt.columns.values.tolist():
    print('Carac:')
    print(datos_falt[carac].value_counts())
    print('')
```

#### Reemplazar datos


Reemplazar los datos de G (Género), EC (Estado Civil) y SP (Próximo pago en mora):

G:
    0=Masculino (M)
    1=Femenino (F)
    2=Otro (O)

EC: 
    0=Soltero
    1=Casado
    2=Otros

SP:
    0=Si
    1=No

```python
df['G'] = df['G'].replace({'M': 0, 'F': 1, 'O': 2})
df['EC'] = df['EC'].replace({'Soltero': 0, 'Casado': 1, 'Otros': 2})
df['SP'] = df['SP'].replace({'Si': 0, 'No': 1})

df.info()
```

#### Imputar datos

##### De EC
 Es variable categórica

```python
mas_freq=df["EC"].value_counts().idxmax()
print('Variables mas frecuente de EC: ', mas_freq)

# imputar
df["EC"].replace(np.nan, mas_freq,inplace=True)

df["EC"]=df["EC"].astype('int64')
#comprobar
df["EC"].info()
```

##### De G
 Es variable numérica

```python
med_g=df["G"].astype("float").mean()
print(f'Media de G: {med_g:.2f}' )

#imputar la media
df["G"]=df["G"].replace(np.nan, med_g)
# comprobar si aun hay faltantes
df["G"].count()
```

#### Comprobar si aún hay datos faltantes

```python
df.info()
```

### Generar Modelos

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,f1_score
from sklearn.model_selection import train_test_split

X=df.drop('SP',axis=1)# matriz entrada
y=df['SP'] # vector salida

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=20)
```

#### 1. RF

```python
from sklearn.ensemble import RandomForestClassifier
import pickle

rf=RandomForestClassifier(n_estimators=50, max_depth=20, random_state=42)
#Entrenar
rf.fit(X_train,y_train)
#Predecir
y_pred=rf.predict(X_test) 
```

##### Validacion para RF

```python
exac=accuracy_score(y_test,y_pred)
prec = precision_score(y_test, y_pred)
sens = recall_score(y_test, y_pred)
print("Matriz de confusion: \n", confusion_matrix(y_test, y_pred))
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
spec = tn / (tn + fp)
f1=f1_score(y_test, y_pred)

print(f"Exactitud RF: {exac:.2f}")
print(f"Precisión RF: {prec:.2f}")
print(f"Sensibilidad RF: {sens:.2f}")
print(f"Especificidad RF: {spec:.2f}")
print(f"F1-score RF: {f1:.2f}")
```

#### 2. XGB

```python
from xgboost import XGBClassifier

xgb=XGBClassifier(use_label_encoder=False,learning_rate=0.1, max_depth=20, n_estimators=50)
#Entrenar
xgb.fit(X_train,y_train)
#prediccion
y_pred=xgb.predict(X_test)
```

##### Validacion para XGBoost

```python
exac=accuracy_score(y_test,y_pred)
prec = precision_score(y_test, y_pred)
sens = recall_score(y_test, y_pred)
print("Matriz de confusion: \n", confusion_matrix(y_test, y_pred))
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
spec = tn / (tn + fp)
f1=f1_score(y_test, y_pred)

print(f"Exactitud XGB: {exac:.2f}")
print(f"Precisión XGB: {prec:.2f}")
print(f"Sensibilidad XGB: {sens:.2f}")
print(f"Especificidad XGB: {spec:.2f}")
print(f"F1-score XGB: {f1:.2f}")
```

### Guardar el mejor modelo

```python
#Guardar el mejor modelo
filename = 'finalized_model.sav'
pickle.dump(xgb, open(filename, 'wb'))

loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
```
